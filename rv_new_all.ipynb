#Part 1: Data Import and Additional Processing


Preprocessing was performed via Cynthia Koopmans code using LabelEncoder and One Hot Encoding.

This processing is to move the label column to the front of the columns.

```python
import numpy as np
import pandas as pd
```

```python
traindata = pd.read_csv("/content/drive/MyDrive/Colab Current Work/enc_traindata.csv")
testdata = pd.read_csv("/content/drive/MyDrive/Colab Current Work/enc_testdata.csv")
```

```python
print(traindata.shape)
traindata.head()
print(traindata.iloc[0:6, 0:4])
```

```python
#move label column to column 1 (column 0 is a unnamed numerical)
move_label_column_train = traindata.pop("label")

# insert column with insert(location, column_name, column_value)

traindata.insert(0, "label", move_label_column_train)

print(traindata.shape)
traindata.head()
print(traindata.iloc[0:6, 0:4])

#move label column to column 1 (column 0 is an unnamed numerical)
move_label_column_test = testdata.pop("label")

testdata.insert(1, "label", move_label_column_test)

print(testdata.shape)
testdata.head()

```

```python
#Change the labels to 0 for normal and 1 for attack

# take label column
labeldf=traindata['label']
labeldf_test=testdata['label']
# change the label column
newlabeldf=labeldf.replace({ 'normal' : 0, 'neptune' : 1 ,'back': 1, 'land': 1, 'pod': 1, 'smurf': 1, 'teardrop': 1,'mailbomb': 1, 'apache2': 1, 'processtable': 1, 'udpstorm': 1, 'worm': 1,
                           'ipsweep' : 1,'nmap' : 1,'portsweep' : 1,'satan' : 1,'mscan' : 1,'saint' : 1
                           ,'ftp_write': 1,'guess_passwd': 1,'imap': 1,'multihop': 1,'phf': 1,'spy': 1,'warezclient': 1,'warezmaster': 1,'sendmail': 1,'named': 1,'snmpgetattack': 1,'snmpguess': 1,'xlock': 1,'xsnoop': 1,'httptunnel': 1,
                           'buffer_overflow': 1,'loadmodule': 1,'perl': 1,'rootkit': 1,'ps': 1,'sqlattack': 1,'xterm': 1})
newlabeldf_test=labeldf_test.replace({ 'normal' : 0, 'neptune' : 1 ,'back': 1, 'land': 1, 'pod': 1, 'smurf': 1, 'teardrop': 1,'mailbomb': 1, 'apache2': 1, 'processtable': 1, 'udpstorm': 1, 'worm': 1,
                           'ipsweep' : 1,'nmap' : 1,'portsweep' : 1,'satan' : 1,'mscan' : 1,'saint' : 1
                           ,'ftp_write': 1,'guess_passwd': 1,'imap': 1,'multihop': 1,'phf': 1,'spy': 1,'warezclient': 1,'warezmaster': 1,'sendmail': 1,'named': 1,'snmpgetattack': 1,'snmpguess': 1,'xlock': 1,'xsnoop': 1,'httptunnel': 1,
                           'buffer_overflow': 1,'loadmodule': 1,'perl': 1,'rootkit': 1,'ps': 1,'sqlattack': 1,'xterm': 1})
# put the new label column back
traindata['label'] = newlabeldf
testdata['label'] = newlabeldf_test
print(traindata['label'].head())
```

```python
#Exports a csv of the data so it can be used again

traindata.to_csv('numlabeled_traindata.csv', index=False)
testdata.to_csv('numlabeled_testdata.csv', index=False)
```

```python
traindata = pd.read_csv("/content/drive/MyDrive/Colab Current Work/numlabeled_traindata.csv")
testdata = pd.read_csv("/content/drive/MyDrive/Colab Current Work/numlabeled_testdata.csv")
```

```python
traindata.head()
```

```python
#change this to correlate with proper locations of new data
X = traindata.iloc[:,1:42]
Y = traindata.iloc[:,0]
C = testdata.iloc[:,0]
T = testdata.iloc[:,1:42]

scaler = Normalizer().fit(X)
trainX = scaler.transform(X)

scaler = Normalizer().fit(T)
testT = scaler.transform(T)


traindata = np.array(trainX)
trainlabel = np.array(Y)

testdata = np.array(testT)
testlabel = np.array(C)



#traindata = X_train
#testdata = X_test
#trainlabel = y_train
#testlabel = y_test
```

```python
#Linear Regression
print("-----------------------------------------LR---------------------------------")
model = LogisticRegression()
model.fit(traindata, trainlabel)

# make predictions
expected = testlabel
np.savetxt('/content/drive/MyDrive/Colab Results/classical/expected.txt', expected, fmt='%01d')
predicted = model.predict(testdata)
proba = model.predict_proba(testdata)

np.savetxt('/content/drive/MyDrive/Colab Results/classical/predictedlabelLR.txt', predicted, fmt='%01d')
np.savetxt('/content/drive/MyDrive/Colab Results/classical/predictedprobaLR.txt', proba)

y_train1 = expected
y_pred = predicted
accuracy = accuracy_score(y_train1, y_pred)
recall = recall_score(y_train1, y_pred , average="binary")
precision = precision_score(y_train1, y_pred , average="binary")
f1 = f1_score(y_train1, y_pred, average="binary")

print("accuracy")
print("%.3f" %accuracy)
print("precision")
print("%.3f" %precision)
print("recall")
print("%.3f" %recall)
print("f1score")
print("%.3f" %f1)
```
